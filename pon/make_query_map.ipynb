```python
data = []

for file in os.listdir('./data/labeled'):
    filename = f"./data/labeled/{file}"
    df = pd.read_csv(filename, index_col=0)
    query_dict = extract_query_from_df(df)
    
    for query_type_key, query_type_value in query_dict.items():
        for query in query_type_value:
            data.append((file, query_type_key, query))

df_result = pd.DataFrame(data, columns=['file', 'query_type', 'query'])
df_result.head(20)
```

```python
df_result.to_csv("./data/labeled/file_query_map.csv", index=False)
```
